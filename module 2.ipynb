```python

import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

```

```python
graph = nx.Graph() 
```

```python

df = pd.read_csv("dreaddit-train.csv")     
            
for _, row in df.iterrows():
    graph.add_edge(row['subreddit'], row['post_id'])
    
 
        
```

```python
nx.write_graphml(graph, "subreddit.graphml")
```

```python
subreddit_counts = df['subreddit'].value_counts()

# Plot the bar graph
plt.figure(figsize=(10, 6))
subreddit_counts.plot(kind='bar', color='blue')
plt.title('Number of Occurrences')
plt.xlabel('Subreddit')
plt.ylabel('Count')

# Show the graph
plt.show()
```
